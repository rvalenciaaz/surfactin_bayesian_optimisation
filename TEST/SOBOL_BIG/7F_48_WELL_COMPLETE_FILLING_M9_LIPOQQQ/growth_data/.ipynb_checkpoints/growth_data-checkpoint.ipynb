```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
master_tables=[pd.read_csv(str(i)+"_master_table.csv") for i in range (0,6)]
```

```python
blank_tables=[pd.read_csv("blank_plate_"+str(i)+".csv") for i in range (0,6)]
```

```python
OD_tables=[pd.read_csv("iteration_"+str(i)+"_run_OD.csv",sep="\t") for i in range (0,6)]
```

```python
master_dictionaries=[dict(zip(j["well_position"],j["replicate"])) for j in master_tables]
```

```python
correction=[j.drop(0).drop("Time [s]",axis=1).mean(axis=0) for j in blank_tables]
```

```python
corrected_OD=[]
for df, s in zip(OD_tables,correction):
    common_columns = df.columns.intersection(s.index)
    df[common_columns] = df[common_columns] - s
    corrected_OD.append(df)
```

```python
OD_hours=[g.assign(Time=g['Time [s]']/3600).drop("Time [s]",axis=1) for g in corrected_OD]
```

```python
OD_samples = [a.rename(columns=b).filter(regex='(_|Time)') for a, b in zip(OD_hours, master_dictionaries)]
```

```python
palette = {
    0: '#0000FF',  # Blue
    1: '#FF0000',  # Red
    2: '#77DD77',  # Pastel Green
    3: '#FDFD96',  # Pastel Yellow
    4: '#800080',  # Purple
    5: '#ADD8E6',  # Light Blue
}

colors = palette

plt.figure(figsize=(10, 6))  # Optional: Specifies the figure size

for index, df in enumerate(OD_samples):
    time_column = df.filter(like='Time').columns[0]  # Assumes first 'Time' like column is the x-axis
    for col in df.columns:
        if col != time_column:  # Avoid plotting the time column against itself
            plt.plot(df[time_column], df[col], label=f'{col}', color=colors[index])

plt.title('Time Series Plots for OD Measurements')
plt.xlabel('Time')
plt.ylabel('Optical Density (OD)')
plt.legend()
plt.grid(True)  # Optional: Adds a grid for better visual guidance
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming OD_samples is your list of DataFrames
# Create an empty DataFrame to collect processed data
all_data = pd.DataFrame()

# Extract data and unify into one DataFrame with necessary labels
for df in OD_samples:
    # Extract the time column
    time_column = df.filter(like='Time').columns[0]
    # Melt other columns to long format
    melted = df.melt(id_vars=[time_column], var_name='Variable', value_name='OD')
    # Parse the variable names into components
    melted[['Batch', 'SampleType', 'Replicate']] = melted['Variable'].str.split('_', expand=True)
    # Append to the collected DataFrame
    all_data = pd.concat([all_data, melted], ignore_index=True)

# Group by Batch, SampleType, and Time to compute mean and std dev
grouped = all_data.groupby(['Batch', 'SampleType', time_column])
mean_data = grouped['OD'].mean().reset_index()
std_dev = grouped['OD'].std().reset_index()

# Compute confidence intervals (95% CI assuming normal distribution)
mean_data['Lower CI'] = mean_data['OD'] - 1.96 * (std_dev['OD'] / np.sqrt(grouped.size().reset_index(name='Counts')['Counts']))
mean_data['Upper CI'] = mean_data['OD'] + 1.96 * (std_dev['OD'] / np.sqrt(grouped.size().reset_index(name='Counts')['Counts']))

# Plotting
plt.figure(figsize=(12, 8))

for (batch, sample_type), group_df in mean_data.groupby(['Batch', 'SampleType']):
    plt.plot(group_df[time_column], group_df['OD'], label=f'{batch}_{sample_type} Mean')
    plt.fill_between(group_df[time_column], group_df['Lower CI'], group_df['Upper CI'], alpha=0.3, label=f'{batch}_{sample_type} 95% CI')

plt.title('Mean and 95% Confidence Interval of OD Measurements by Batch and Sample Type')
plt.xlabel('Time')
plt.ylabel('Optical Density (OD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assume OD_samples is your list of DataFrames
# Create an empty DataFrame to collect processed data
all_data = pd.DataFrame()

# Extract data and unify into one DataFrame with necessary labels
for df in OD_samples:
    # Extract the time column
    time_column = df.filter(like='Time').columns[0]
    # Melt other columns to long format
    melted = df.melt(id_vars=[time_column], var_name='Variable', value_name='OD')
    # Parse the variable names into components
    melted[['Batch', 'SampleType', 'Replicate']] = melted['Variable'].str.split('_', expand=True)
    # Append to the collected DataFrame
    all_data = pd.concat([all_data, melted], ignore_index=True)

# Group by Batch, SampleType, and Time to compute mean and std dev
grouped = all_data.groupby(['Batch', 'SampleType', time_column])
mean_data = grouped['OD'].mean().reset_index()
std_dev = grouped['OD'].std().reset_index()

# Compute confidence intervals (95% CI assuming normal distribution)
mean_data['Lower CI'] = mean_data['OD'] - 1.96 * (std_dev['OD'] / np.sqrt(grouped.size().reset_index(name='Counts')['Counts']))
mean_data['Upper CI'] = mean_data['OD'] + 1.96 * (std_dev['OD'] / np.sqrt(grouped.size().reset_index(name='Counts')['Counts']))

# Define the color palette for each batch
palette = {
    '0': '#0000FF',  # Blue
    '1': '#FF0000',  # Red
    '2': '#77DD77',  # Pastel Green
    '3': '#FDFD96',  # Pastel Yellow
    '4': '#800080',  # Purple
    '5': '#ADD8E6',  # Light Blue
}

# Plotting
plt.figure(figsize=(12, 8))

legend_labels = set()  # To keep track of legend entries
for (batch, sample_type), group_df in mean_data.groupby(['Batch', 'SampleType']):
    color = palette.get(batch, '#000000')  # Default to black if batch not found
    label = f'Batch {batch}'
    if label not in legend_labels:
        plt.plot(group_df[time_column], group_df['OD'], label=label, color=color)
        plt.fill_between(group_df[time_column], group_df['Lower CI'], group_df['Upper CI'], color=color, alpha=0.3)
        legend_labels.add(label)
    else:
        plt.plot(group_df[time_column], group_df['OD'], color=color)
        plt.fill_between(group_df[time_column], group_df['Lower CI'], group_df['Upper CI'], color=color, alpha=0.3)

plt.title('Mean and 95% Confidence Interval of OD Measurements by combination or controls')
plt.xlabel('Time')
plt.ylabel('Optical Density (OD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming OD_samples is your list of DataFrames
# Create an empty DataFrame to collect the last value data
last_values_data = pd.DataFrame()

# Extract data and unify into one DataFrame with necessary labels
for df in OD_samples:
    # Extract the time column
    time_column = df.filter(like='Time').columns[0]
    # Take the last time point
    last_time_df = df.loc[df[time_column] == df[time_column].max()]
    # Melt other columns to long format
    melted = last_time_df.melt(id_vars=[time_column], var_name='Variable', value_name='OD')
    # Parse the variable names into components
    melted[['Batch', 'SampleType', 'Replicate']] = melted['Variable'].str.split('_', expand=True)
    # Append to the collected DataFrame
    last_values_data = pd.concat([last_values_data, melted], ignore_index=True)

# Now create the plot
palette = {
    '0': '#0000FF',  # Blue
    '1': '#FF0000',  # Red
    '2': '#77DD77',  # Pastel Green
    '3': '#FDFD96',  # Pastel Yellow
    '4': '#800080',  # Purple
    '5': '#ADD8E6',  # Light Blue
}

# Plotting
plt.figure(figsize=(14, 8))

# Group data by Batch and SampleType for plotting
grouped = last_values_data.groupby(['Batch', 'SampleType'])

# To handle colors and labels correctly in the legend
patches = []  # To handle custom legends for colors

for (batch, sample_type), group_df in grouped:
    # Use the batch to determine the color
    color = palette.get(batch, '#000000')  # Default to black if batch not found
    # Plot the boxplot for each group
    box = plt.boxplot(group_df['OD'], positions=[len(patches)], patch_artist=True, boxprops=dict(facecolor=color),widths=0.7)
    patches.append(box['boxes'][0])  # Add the patch object for the legend

plt.xticks(range(len(patches)), [f'{batch}_{sample_type}' for batch, sample_type in grouped.groups.keys()], rotation=90,fontsize=16)
plt.yticks(fontsize=16)
plt.title('Boxplot of the Last OD Measurement for Each Batch/Sample Type')
plt.ylabel('Optical Density (OD)')
#plt.grid(True)

# Create custom legend
from matplotlib.patches import Patch
legend_colors = [Patch(facecolor=palette[batch], label=f'Batch {batch}') for batch in sorted(palette.keys())]
plt.legend(handles=legend_colors)

# Saving the plot
#save_path = 'surfactinc_concentration_boxplot_ordered.png'
#plt.savefig(save_path, dpi=600, transparent=True)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming OD_samples is your list of DataFrames
# Create an empty DataFrame to collect the last value data
last_values_data = pd.DataFrame()

# Extract data and unify into one DataFrame with necessary labels
for df in OD_samples:
    # Extract the time column
    time_column = df.filter(like='Time').columns[0]
    # Take the last time point
    last_time_df = df.loc[df[time_column] == df[time_column].max()]
    # Melt other columns to long format
    melted = last_time_df.melt(id_vars=[time_column], var_name='Variable', value_name='OD')
    # Parse the variable names into components
    melted[['Batch', 'SampleType', 'Replicate']] = melted['Variable'].str.split('_', expand=True)
    # Append to the collected DataFrame
    last_values_data = pd.concat([last_values_data, melted], ignore_index=True)

# Create a dictionary to hold colors for each batch
palette = {
    '0': '#0000FF',  # Blue
    '1': '#FF0000',  # Red
    '2': '#77DD77',  # Pastel Green
    '3': '#FDFD96',  # Pastel Yellow
    '4': '#800080',  # Purple
    '5': '#ADD8E6',  # Light Blue
}

# Calculate means and sort groups by these means
means = last_values_data.groupby(['Batch', 'SampleType']).OD.mean()
sorted_groups = means.sort_values().index

plt.figure(figsize=(14, 8))

# Plot each group in the sorted order
positions = []
labels = []
for i, (batch, sample_type) in enumerate(sorted_groups):
    group_df = last_values_data[(last_values_data['Batch'] == batch) & (last_values_data['SampleType'] == sample_type)]
    color = palette.get(batch, '#000000')  # Default to black if batch not found
    box = plt.boxplot(group_df['OD'], positions=[i], patch_artist=True, boxprops=dict(facecolor=color), widths=0.7)
    # Plotting the mean
    mean_value = group_df['OD'].mean()
    plt.scatter([i], [mean_value], color='black', zorder=5)
    positions.append(i)
    labels.append(f'{batch}_{sample_type}')

plt.xticks(positions, labels, rotation=90, fontsize=16)
plt.yticks(fontsize=16)
plt.title('Boxplot of the Last OD Measurement (24 hrs) for Each Batch/Sample Type Ordered by Mean')
plt.ylabel('Optical Density (OD)',fontsize=16)

# Create custom legend
from matplotlib.patches import Patch
legend_colors = [Patch(facecolor=palette[batch], label=f'Batch {batch}') for batch in sorted(palette.keys())]
plt.legend(handles=legend_colors)
plt.tight_layout()

save_path = 'OD_boxplot_ordered.png'
plt.savefig(save_path, dpi=600, transparent=True,bbox_inches="tight")

plt.show()
```
